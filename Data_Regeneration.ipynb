```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Load and process the first dataset
data = pd.read_csv("al_transac_data_count_fee.csv", header=None)
df1 = data[1]

# Load and process the second dataset
data2 = pd.read_csv("daily_transactions.csv", header=None)
df2 = data2[1].replace(0, np.nan)
df2.dropna(inplace=True)
df2 = df2[1:]
df2 = df2.astype(float)

# Create a figure and axis object
fig, ax = plt.subplots(2, figsize=[20, 15], sharex=False)

# Plot the two data series on the same axis, specifying different colors
df1.plot(ax=ax[0], label='Dataset 1', color='blue')  # Specify color for first line
df2.plot(ax=ax[1], label='Dataset 2', color='red')   # Specify color for second line

# Add legend
# fig.legend()

# Set the y-axis to use scientific notation
# ax[0].ticklabel_format(style='sci', scilimits=(0,0), axis='y')
# ax[1].ticklabel_format(style='sci', scilimits=(0,0), axis='y')
ax[0].ticklabel_format(style='plain', axis='y')
ax[1].ticklabel_format(style='plain', axis='y')

# Set the y-axis label
ax[0].set_ylabel("Daily Transaction", fontsize=20)
ax[0].set_xlabel("Algorand(Days After Genesis)", fontsize=20)
ax[1].set_ylabel("Daily Transaction", fontsize=20)
ax[1].set_xlabel("Beacon Chain(Days After Genesis)", fontsize=20)

ax[0].tick_params(axis='both', labelsize=20)
ax[1].tick_params(axis='both', labelsize=20)
# ax[1].set_xlim(pd.Timestamp('2015-07-29'), pd.Timestamp('2023-09-04'))
# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df3 = pd.read_csv('avg_blk_time.csv')

# Convert Unix timestamps to datetime
df3['Timestamp'] = pd.to_datetime(df3['Timestamp'], unit='ms')

# Set the timestamp as the index
# df3.set_index('Timestamp', inplace=True)

# Create the plot
fig, ax = plt.subplots(figsize=[20, 15])

# Plot the 'Value' column
df3['Value'].plot(ax=ax, color='blue')

# Set labels and title
ax.set_ylabel("Daily Block Time", fontsize=20)
ax.set_title("Daily Block Time Over Time", fontsize=22)
ax.tick_params(axis='both', which='major', labelsize=20)

# Draw a horizontal line at y=3
ax.axhline(y=3, color='red', linestyle='--', linewidth=2)

# Set x-axis limits
ax.set_xlim(pd.Timestamp('2019-08'), pd.Timestamp('2021-09'))

# Display the plot
plt.show()
```
